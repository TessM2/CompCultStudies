**Uploading files into colab**

Let's begin with a quick lesson on one of the few essential pieces of info we haven't covered yet: how to upload text files (or other files) into colab. As I mentioned in previous weeks, one of the few differences between colab (which we're using) and Jupyter notebooks (which a lot of the text book materials you'll be working from use) is how they handle the process of uploading in text or other files - and also downloading them back, again, to your computer. Indeed, when you read textbooks (like Mel Walsh's) that show you code cells which involve the process of uploading in files, the code they present won't work in colab. Here we'll assume we're uploading a text file, but later on we'll start loading in spreadsheet style files, in which case we'll change the code slightly at the last line (and we'll talk about that later).

```python
#loading a text file into colab

from google.colab import files

# Upload the file
uploaded = files.upload()

# Get the filename (assuming only one file is uploaded)
filename = list(uploaded.keys())[0]

# Read the contents of the file
with open(filename, 'r', encoding='utf-8') as f:
    myfilecontents = f.read()
```

```python
print(myfilecontents)
```

```python
import re
from collections import Counter

def split_into_words(any_chunk_of_text):
    lowercase_text = any_chunk_of_text.lower()
    split_words = re.split("\W+", lowercase_text)
    return split_words

filepath_of_text = "../texts/literature/The-Yellow-Wallpaper_Charlotte-Perkins-Gilman.txt"
number_of_desired_words = 40

stopwords = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours',
 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers',
 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves',
 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are',
 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does',
 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until',
 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into',
 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down',
 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here',
 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more',
 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so',
 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now', 've', 'll', 'amp']

full_text = open(filepath_of_text, encoding="utf-8").read()

all_the_words = split_into_words(full_text)
meaningful_words = [word for word in all_the_words if word not in stopwords]
meaningful_words_tally = Counter(meaningful_words)
most_frequent_meaningful_words = meaningful_words_tally.most_common(number_of_desired_words)

most_frequent_meaningful_words
```

First, let's discuss the database. It's a grid, or a dataframe, and it's the primary way we organize and work with data - even, often, textual data - in python. It can be saved in an excel file, a csv, or other formats. Let's quickly discuss them. Then let's take a tiny peak at what you'll be working on for next week, loading files into work with pandas.

Pandas is a package we use to work with grid style data in what we call a "dataframe". To load in this package, we write this line of code:

```python
import pandas as pd
```

Note, you could actually call pandas anything you want, after the "as" (the "as" means you're giving the package that name), like naming a variable. Almost everyone calls pandas pd though, so I'd import it as pd.

What pandas helps you do is work with spreadsheet style data in the form of what's called a "dataframe"; so, you can use pandas to load in the data from, say, an excel sheet, or a csv file, and then you can manipulate that data with pandas. Here's a catch, though: just like colab has a special way to load in a text file, different from jupyter notebook, it also has a different way to load in an excel file or csv, or other file with grid like data, to name it as a dataframe in pandas. So, if you know nothing else to work on the homework — working through the pandas lesson in melanie's tectbook - you should know that the way she loads in dataframes to pandas is not how you'll eventually do it. Here's how you'll do it:

To load in an excel file:

```python

#note, again, you don't need to reload in these packages, likepandas, since we already did; but I'm giving you the code again to do so in case you copy and paste this elsewhere
from google.colab import files
import pandas as pd

# Upload the file manually
uploaded = files.upload()

# Get the filename (assuming one file is uploaded)
filename = list(uploaded.keys())[0]

# Load the Excel file into a Pandas DataFrame
df = pd.read_excel(filename)

# Display the first few rows of the DataFrame
df.head()
```

To load in a csv file:

```python
from google.colab import files
import pandas as pd

# Upload the CSV file manually
uploaded = files.upload()

# Get the filename (assuming only one file is uploaded)
filename = list(uploaded.keys())[0]

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(filename)

# Display the first few rows of the DataFrame
df.head()
```
