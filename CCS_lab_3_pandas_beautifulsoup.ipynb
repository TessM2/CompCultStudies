Let's start with a quick review of one last coding topic. During the past few weeks, we learned how to do things like make for loops and conditional statements. And we learned to write this type of code in formats on multiple lines. this same code, however, can also be written in neater formats, that take up only one line. But since you will often encounter these neater formats when reading other people's (or bots) code, I just want to help you recognize them.

Let's start with a list comprehension. This is one line of code for when we want to loop through a list and do an operation for every item in the list. Thus far, we've created a for loop, for that operation, like this:

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = []

for num in numbers:
    squared_numbers.append(num ** 2)

print(squared_numbers)
```

But people often put this process on one line as a "list comprehension", like this:

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = [num ** 2 for num in numbers]

print(squared_numbers)
```

We can also put conditional statements on one line. What do you think this does?

```python
age = 18
status = "Adult" if age >= 18 else "Minor"

print(status)
```

That's called a "ternary operator" putting the conditional on one line. How would you write this same code in the multiline format you know?

Here's another type of way to put a process you might put on multiple lines on one line. It looks a lot like the list comprehension, and seems to do the same thing. But looks different. what is it? Ask chat gpt (or google) about it. what did you find out?

```python
numbers = [1, 2, 3, 4, 5]
squared_gen = (num ** 2 for num in numbers)

# Print values one by one
for value in squared_gen:
    print(value)
```

What beautiful soup will help us do is parse or read (or alter) a .xml file with tei encoding. So, let's first upload in the xml file of that text we've been looking at.

```python
from google.colab import files

# Upload the XML file
uploaded = files.upload()

# Replace 'your_file.xml' with the actual filename you uploaded
xml_filename = list(uploaded.keys())[0]  # Get the uploaded file name dynamically

# Read the XML file
with open(xml_filename, "r", encoding="utf-8") as file:
    xml_content = file.read()

```

```python

# Parse the XML using BeautifulSoup
soup = BeautifulSoup(xml_content, "xml")

# Print the first 500 characters of the formatted XML
print(soup.prettify()[:10000])
```

```python
#let's say we wanted to just get rid of all the xml, what would we do?
```

```python
# Extract only the text, removing XML tags
plain_text = soup.get_text()

# Print the first 1000 characters to preview
print(plain_text[:10000])
```

OK, we can always do that. But maybe we want to use the labels. Maybe, for example, we want to use the tags for all physical locations. Here's how to search through the xml and locate elements in tags:

```python
place_names = []  # Create an empty list

# Loop through all <placeName> tags
for place in soup.find_all("placeName"):
    place_text = place.get_text(strip=True)  # Extract text and strip spaces
    place_names.append(place_text)  # Add to the list

print(place_names)  # Print the final list
```

Even if we don't want to use the xml tags on specific words, it still might be useful to use someof the tags - like, say, the headers, to organize the data into a data frame. For example, let's say I wanted to make a dataframe with the title of this chapter (in header one) and then rows for the text of each part of the chapter. I could...

```python
#first check out all the headers
headings = [h.get_text() for h in soup.find_all("head")]

# Print all extracted headings
for i, heading in enumerate(headings, 1):
    print(f"Heading {i}: {heading}")
```

```python
#Then make a dataframe - you dont need to understand all this now; it's just to show you how and in what ways we can parse xml. You'll do more parsing for homework (except not of xml, of html!)
import pandas as pd
#note we already imported pandas in the earlier lesson but a reminder in case we hadn't

# Create an empty list to store extracted data
data = []

# Extract the first <head> tag (this is the chapter title)
chapter_title = soup.find("head").get_text(strip=True)

# Find all section divisions <div2> (these contain "Part I, Part II, etc.")
for div in soup.find_all("div2"):
    # Extract part name (the <head> inside the <div2>)
    part_title = div.find("head").get_text(strip=True) if div.find("head") else "Unknown Part"

    # Extract text inside this part (excluding XML tags)
    part_text = "\n".join(p.get_text() for p in div.find_all("p"))

    # Append data to the list
    data.append(["Civilization in the United States", "Matthew Arnold", chapter_title, part_title, part_text])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Book Title", "Author", "Chapter Title", "Part", "Text"])

# Display the DataFrame
df.head()

```

***one key point to note about html, is that text is almost always in the paragraph tag, p (there was also a p tag for paragraphs in tei/xml)***

For homework, this code block, grapping the HTML from a wiki page and then examining it with beautiful soup might be helpful. It also then locates the text within the p tags.

```python
#import requests
import requests
#get the wiki page html
response = requests.get("https://en.wikipedia.org/wiki/Banana")
html_string = response.text

```

```python
#parse with beautfiul soup
document = BeautifulSoup(html_string, "html.parser")
print(document)
```

```python
#locate everything with the p (paragraph) tags and grab
paragraphs = document.find_all("p")
```

```python
print(paragraphs)
```

```python

# Remove all HTML tags from the extracted paragraphs
clean_paragraphs = [p.get_text(strip=True) for p in paragraphs]

# Print the first few cleaned paragraphs
for para in clean_paragraphs:
    print(para)

#note we'd still have some cleaning to do!
```
