So, here's an example of "getting our data"; we're going to download the story from Project Gutenberg. Don't worry, you're not meant to understand this code, at this point.

```python
#Downloading The Yellow Wall Paper

# Import the required library
import requests

# URL of the text file from Project Gutenberg
url = "https://www.gutenberg.org/files/1952/1952-0.txt"

# Send a GET request to fetch the text file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the text to a variable
    text = response.text

    # Save the text to a file in Colab
    with open("gutenberg_text.txt", "w", encoding="utf-8") as file:
        file.write(text)

    print("Text downloaded and saved as 'gutenberg_text.txt'")
else:
    print(f"Failed to download text. Status code: {response.status_code}")
```

```python
#Now let's check out the Text

# Open and read the downloaded file
with open("gutenberg_text.txt", "r", encoding="utf-8") as file:
    content = file.read()

# Display the first 500 characters of the text
print(content[:500])
```

```python
#Let's display the whole thing:
print(content)
```

**Cleaning the Data**

OK, so let's say we've decided, for whatever reason, this is the appropriate "data" for us to use: we know all about this Project Gutenberg text, and we find it to be suitable for our purposes. The next thing we do is "clean" our data. Again, we'll talk about this much more: but what "cleaning" means is putting the data in a suitable format for the kind of analysis we want to do. So, for example, in this case, let's say we want to run analysis on the story itself. Well, we might have to "clean" some things out of our text, to be sure we don't include elements we don't want. What might we want to "clean" out?

That's right...a lot of things you said. But especially, the one of you who said, the part where it says "Start of Project Gutenberg e-Book." We don't want to make the mistake of including that in our analysis of the story itself.

So, a small example of what our data cleaning process might look like—one tiny piece of that larger process—might be: let's get rid of that.

```python
#Run code to remove the Gutenberg related header from the data (there's actually also a footer! but we'll pass over that here)

# The string to remove
string_to_remove = "*** START OF THE PROJECT GUTENBERG EBOOK 1952 ***"

# Remove the string from the content
updated_content = content.replace(string_to_remove, "")

# Display a preview of the updated text
print("Preview of updated content:")
print(updated_content[:500])

```

**Analysis (and critical iteration)**

OK, now let's say our data is all cleaned up. Now it is time to run our analysis. Remember, a huge, pre-emptive part of this process, on an intellectual level, is considering what type of data analysis can actually answer our question. And ideally, we've done a lot of that thinking; but some of that thinking might also happen in the process of actuallty doing our analysis, as we realize mistakes we made, new complications that emerge, and so on. The process of converting a question we have into an actual process of analysis is called "operationalizing" that question; and it involves a lot of thought—and knowledge of the methods available (not to mention a healthy dose of self-skepticism/criticism, etc.) But let's, again, simplify for now, by assuming this (brief) line of reasoning:

We had a question about "objects" in these stories and their relation to people. And the first most basic question we had was: what objects are described? and how are they described? Well, we might think about processes for answering each of those questions. But to answer the first, in this very simply example, we might just say: well, let's locate all the nouns in the story, and take a look at them. Because most objects are referred to by nouns, this might give us an idea of the objects present.

I bet you're already coming up with a lot of good objections. Good, that's how it should be—that's how we make these processes better; or, how we decide not to use them at all. It's worth noting, though, that if we want to analyze quite a large number of stories—instead of one—then "close" reading them might not be an option (or might it, actually? always good to think about—would close reading a smaller sample actually be better than "distant reading" an entire giant corpus? questions we will, and must, consider)

But for now, let's locate those nouns:


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, FreqDist

# Ensure the necessary NLTK resources are downloaded
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

# Tokenize the text
tokens = word_tokenize(updated_content)

# Perform Part-of-Speech tagging
pos_tags = pos_tag(tokens)

# Extract nouns (NN, NNS, NNP, NNPS)
nouns = [word for word, pos in pos_tags if pos in ['NN', 'NNS', 'NNP', 'NNPS']]

# Calculate frequency distribution of nouns
noun_freq = FreqDist(nouns)

# Get the top 50 nouns by frequency
top_50_nouns = noun_freq.most_common(50)

# Print the top 50 nouns
print("Top 50 nouns by frequency:")
for noun, freq in top_50_nouns:
    print(f"{noun}: {freq}")
```

**Critically Assess, Rinse, Repeat**

There's a lot of problems here. What are they? How might we fix them? AS you implement procedures, you need to check the results and figure out what's going wrong, and then fix those problems. Some problems will be more technical: e.g., why are we seeing all this punctuation? do we need to do something more to "clean" our text? Others are more conceptual: well, are nouns and objects the same? Is there a way we can target "objects" or even physical objects better? When we get farther along in the course, we'll have some better answers to that question (the last one). For now though, this is just to say: we've gotta fix things! Let's try to, for now, just eliminate the issue of those random punctuation marks and single letter's (from contractions) by taking a few extra "cleaning" measures towards those ends.

```python
# Remove punctuation from the text
updated_content_no_punctuation = updated_content.translate(str.maketrans("", "", string.punctuation))

# Tokenize the text
tokens = word_tokenize(updated_content_no_punctuation)

# Perform Part-of-Speech tagging
pos_tags = pos_tag(tokens)

# Extract nouns (NN, NNS, NNP, NNPS) and filter out unwanted tokens
nouns = [
    word for word, pos in pos_tags
    if pos in ['NN', 'NNS', 'NNP', 'NNPS'] and word.isalpha() and len(word) > 1
]

# Calculate frequency distribution of nouns
noun_freq = FreqDist(nouns)

# Get the top 50 nouns by frequency
top_50_nouns = noun_freq.most_common(50)

# Print the top 50 nouns
print("Top 50 nouns by frequency:")
for noun, freq in top_50_nouns:
    print(f"{noun}: {freq}")
```

**Present**

Once all that is done, the last step is to present! This is the step of data storytelling; and it's our job to do it honeslty and responsibly instead of "lying with statistics" to the best of our ability. Part of this process though might be visualizing our data; so let's make a little chart presentoung our results

```python
import matplotlib.pyplot as plt

# Assuming the frequency list is saved in `top_50_nouns`
# Extract the top 10 nouns and their frequencies
top_10_nouns = top_50_nouns[:10]
nouns = [noun for noun, freq in top_10_nouns]
frequencies = [freq for noun, freq in top_10_nouns]

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(nouns, frequencies)
plt.xlabel('Nouns', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Top 10 Nouns by Frequency', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
